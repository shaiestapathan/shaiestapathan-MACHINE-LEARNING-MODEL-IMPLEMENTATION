```

# âœ… CodTech Internship â€“ Task 4: Spam Email Detection using Naive Bayes Classifier

# ğŸ“¦ Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# ğŸ¨ Styling
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

print("ğŸ“ Loading Dataset...")
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv", sep='\t', names=["label", "message"])
print(f"âœ”ï¸ Dataset Loaded: {df.shape[0]} messages")

# ğŸ“Š Initial Data Analysis
spam_count = df[df.label == 'spam'].shape[0]
ham_count = df[df.label == 'ham'].shape[0]
print(f"ğŸ“Œ Spam: {spam_count} | Ham: {ham_count}")

# Plot Class Distribution
sns.countplot(x="label", data=df, palette="Set2")
plt.title("ğŸ“Š Distribution of Messages: Spam vs Ham")
plt.xlabel("Message Type")
plt.ylabel("Count")
plt.show()

# ğŸ§¹ Data Preprocessing
df['label_num'] = df.label.map({'ham':0, 'spam':1})
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['message'])
y = df['label_num']

# ğŸ”€ Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"ğŸ”§ Training Set: {X_train.shape[0]} | Test Set: {X_test.shape[0]}")

# ğŸ¤– Model Training - Naive Bayes
model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# ğŸ“ˆ Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"âœ… Model Accuracy: {round(accuracy * 100, 2)}%")
print("\nğŸ“‹ Classification Report:")
print(classification_report(y_test, y_pred))

# ğŸ” Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='YlGnBu', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.title("ğŸ“Œ Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ğŸ“ Final Summary
print("ğŸ§¾ Final Summary")
print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
print(f"â€¢ Total Messages: {df.shape[0]}")
print(f"â€¢ Algorithm Used: Multinomial Naive Bayes")
print(f"â€¢ Vectorization: CountVectorizer")
print(f"â€¢ Accuracy Score: {round(accuracy * 100, 2)}%")
print(f"â€¢ Minimal false positives and false negatives observed.")

print("\nğŸ“Œ Key Insights:")
print("â€¢ Excellent baseline performance with basic preprocessing.")
print("â€¢ Naive Bayes is fast, interpretable, and effective for text classification.")
print("â€¢ Enhancements can include TF-IDF, stopwords removal, and advanced models like SVM.")

print("\nğŸ¯ Conclusion:")
print("This project demonstrates a high-performing spam detection model using Scikit-learn.")
print("With further tuning, it can be deployed in real-world email filtering systems.")

print("\nâœ… Task 4 Successfully Completed!")
```
